# Data Finder

Simple set of plots to help find MP and towyo data.

```python
import numpy as np
import glob
import munch
import utils
import matplotlib.pyplot as plt
```

# Towyos & MPs

Load towyos and MP data.

```python
data_dir = "proc_data"

# Load processed tow yos
data_files = glob.glob("../" +data_dir+ "/TY*.mat")
TYs = np.asarray([munch.munchify(utils.loadmat(file)) for file in data_files])

# Load MPs
MMP5 = munch.munchify(utils.loadmat('../' +data_dir+ '/M5.mat'))
MMP6 = munch.munchify(utils.loadmat('../' +data_dir+ '/M6.mat'))
PMP1 = munch.munchify(utils.loadmat('../' +data_dir+ '/P1.mat'))
PMP3 = munch.munchify(utils.loadmat('../' +data_dir+ '/P3.mat'))
PMP4 = munch.munchify(utils.loadmat('../' +data_dir+ '/P4.mat'))
PMP5 = munch.munchify(utils.loadmat('../' +data_dir+ '/P5.mat'))
TMP1 = munch.munchify(utils.loadmat('../' +data_dir+ '/T1.mat'))
TMP2 = munch.munchify(utils.loadmat('../' +data_dir+ '/T2.mat'))
TMP3 = munch.munchify(utils.loadmat('../' +data_dir+ '/T3.mat'))
TMP4 = munch.munchify(utils.loadmat('../' +data_dir+ '/T4.mat'))
TMP5 = munch.munchify(utils.loadmat('../' +data_dir+ '/T5.mat'))
TMP6 = munch.munchify(utils.loadmat('../' +data_dir+ '/T6.mat'))
TMP7 = munch.munchify(utils.loadmat('../' +data_dir+ '/T7.mat'))
TMP8 = munch.munchify(utils.loadmat('../' +data_dir+ '/T8.mat'))
TMP9 = munch.munchify(utils.loadmat('../' +data_dir+ '/T9.mat'))
TMP10 = munch.munchify(utils.loadmat('../' +data_dir+ '/T10.mat'))
TMP11 = munch.munchify(utils.loadmat('../' +data_dir+ '/T11.mat'))
TMP12 = munch.munchify(utils.loadmat('../' +data_dir+ '/T12.mat'))
MPs = [MMP5, MMP6, PMP1, PMP3, PMP4, PMP5, TMP1, TMP2, TMP3, TMP4, TMP5,
       TMP6, TMP7, TMP8, TMP9, TMP10, TMP11, TMP12]
```

## P5 Data

```python
P5TYs = [TY for TY in TYs if ((TY.sill == "P5") and (TY.year == 2014))]

P5MPs = [MMP6, PMP5, TMP1, TMP2, TMP3, TMP4, TMP5, TMP6]

fig, ax = plt.subplots(1, 1, figsize=(8, 8))
ax.set_aspect("equal")

for MP in P5MPs:
    ax.plot(MP.lon, MP.lat, 'ro')
    ax.annotate(MP.name,( MP.lon, MP.lat), color='k')
    
for TY in P5TYs:
    ax.plot(TY.mlon, TY.mlat)
    ax.annotate(TY.name, xy=(TY.mlon[-1], TY.mlat[-1]))
```

## P2 Data

```python
P2TYs = [TY for TY in TYs if ((TY.sill == "P2") and (TY.year == 2014))]

P2MPs = [MMP5, PMP1, TMP9, TMP10, TMP11, TMP12]

fig, ax = plt.subplots(1, 1, figsize=(8, 8))
ax.set_aspect("equal")

for MP in P2MPs:
    ax.plot(MP.lon, MP.lat, 'ro')
    ax.annotate(MP.name,( MP.lon, MP.lat), color='k')
    

for TY in P2TYs:
    ax.plot(TY.mlon, TY.mlat)
    ax.annotate(TY.name, xy=(TY.mlon[-1], TY.mlat[-1]))
```
